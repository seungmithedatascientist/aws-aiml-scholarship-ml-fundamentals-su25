## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 2: Train, validation and test

Let's train our transfer learning model! Let's start defining the hyperparameters:

```python
batch_size = 64  # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.2  # fraction of the training data to reserve for validation
num_epochs = 50  # number of epochs for training
num_classes = 50  # number of classes. Do not change this
learning_rate = 0.001  # Learning rate for SGD (or Adam)
opt = 'adam'      # optimizer. 'sgd' or 'adam'
weight_decay = 0.0 # regularization. Increase this to combat overfitting
```

```python
from src.data import get_data_loaders
from src.optimization import get_optimizer, get_loss
from src.train import optimize
from src.transfer import get_model_transfer_learning

# Get a model using get_model_transfer_learning. Use one of the names reported here:
# https://pytorch.org/vision/0.10/models.html
# For example, if you want to load ResNet 18, use "resnet18"
# NOTE: use the hyperparameters defined in the previous cell, do NOT copy/paste the
# values
model_transfer = get_model_transfer_learning(model_name="resnet18", n_classes=num_classes) # YOUR CODE HERE

# train the model
data_loaders = get_data_loaders(batch_size=batch_size)
optimizer = get_optimizer(
    model_transfer,
    learning_rate=learning_rate,
    optimizer=opt,
    weight_decay=weight_decay,
)
loss = get_loss()

optimize(
    data_loaders,
    model_transfer,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/model_transfer.pt",
    interactive_tracking=True
)
```

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/>  __Answer:__ 

For the transfer learning approach, I selected ResNet18 as the backbone architecture because it has been pretrained on ImageNet, which contains millions of images across 1000 categories, giving it robust feature extraction capabilities for visual recognition tasks. Also, I strategically chose to freeze all pretrained layers except the final classification layer because the early and middle layers of ResNet18 have already learned to detect edges, textures, shapes, and high-level visual patterns that are universally useful for image recognition. By keeping these frozen, it can keep this trained knowledge while preventing overfitting on our relatively smaller landmark dataset.

I think this architecture is suitable for landmark classification because landmarks often have similar visual elements to ImageNet categories: Buildings have architectural features similar to houses and towers in ImageNet, natural landmarks share characteristics with mountains and landscapes, and monuments have structural elements comparable to various objects in the pretrained dataset. By replacing only the final fully connected layer with a new one mapping from 512 features to the current 50 landmark classes, the model creates a highly efficient learning setup where the model only needs to learn how to combine the already-extracted high-level features into landmark-specific classifications. This approach dramatically reduces the number of trainable parameters from millions to just around 25,000 (512 × 50), making training faster, more stable, and less vulnerable to overfitting. I expect the transfer learning approach should achieve significantly better accuracy than training from scratch, likely reaching 70-90% accuracy compared to the ~50% achieved with the custom CNN, while requiring fewer training epochs and being more robust to the limited size of the landmark dataset.

## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 4: Export using torchscript

Now, just like we did with our original model, we export the best fit model using torchscript so that it can be used in our application:

```python
from src.predictor import Predictor
from src.helpers import compute_mean_and_std

# First let's get the class names from our data loaders
class_names = data_loaders["train"].dataset.classes

# Then let's move the model_transfer to the CPU
# (we don't need GPU for inference)
model_transfer = model_transfer.cpu()
# Let's make sure we use the right weights by loading the
# best weights we have found during training
# NOTE: remember to use map_location='cpu' so the weights
# are loaded on the CPU (and not the GPU)
model_transfer.load_state_dict(
    torch.load("checkpoints/model_transfer.pt", map_location="cpu")
)

# Let's wrap our model using the predictor class
mean, std = compute_mean_and_std()
predictor = Predictor(model_transfer, class_names, mean, std).cpu()

# Export using torch.jit.script
scripted_predictor = torch.jit.script(predictor)
scripted_predictor.save("checkpoints/transfer_exported.pt")
```
