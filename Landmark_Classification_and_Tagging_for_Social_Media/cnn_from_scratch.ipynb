<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question:** Describe your chosen procedure for preprocessing the data. 
- How does your code resize the images (by cropping, stretching, etc)?  What size did you pick for the input tensor, and why?
- Did you decide to augment the dataset?  If so, how (through translations, flips, rotations, etc)?  If not, why not?

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> **Answer**: 

- My preprocesing pipeline uses a two-step resize and crop approach, where I resize images to 256x256 pixels, and crop images to 224 x 224 pixels. Specifically, for the training set, it randomly selects a 224x224 region; for the validation and test sets, it takes the center 224x224 region. This approach is to avoid stretching and maintain the original aspect ratio of importance features. By resizing to 256 first, there's enough image content to crop from while not being too computationally expensive. 
- I chose 224x224x3 as the input tensor size because:
    - It's a standard size used by many successful CNN architectures (VGG, ResNet, etc.)
    - Large enough to capture important landmark details and features
    - Small enough to allow reasonable batch sizes in GPU memory
    - The square format works well with CNN architectures that use successive pooling layers
- I implemented data augmentation for the training set with RandomCrop(224) that provides translation augmentation by randomly selecting different 224x224 regions from the 256x256 resized image, and RandomHorizontalFlip() that adds horizontal flipping with 50% probability because:
    - Horizontal flips: Most landmarks look natural when flipped horizontally, and this effectively doubles the training data
    - Random cropping: Helps the model become invariant to small translations and focuses on different parts of landmarks
    - No rotation: I avoided rotation because landmarks often have a canonical orientation (buildings, monuments) and rotating them might create unrealistic training examples

<img src="static_images/icons/noun-question-mark-869751.png" alt="?" style="width:25px"/> **Question**: Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

<img src="static_images/icons/noun-answer-3361020.png" alt=">" style="width:25px"/> __Answer:__ 

I began with a VGG-inspired design because VGG networks are commonly known for their simplicity and effectiveness, and the pattern of stacking 3×3 convolutions is proven to work well.

First, I created 5 convolutional blocks with increasing depth:
- Block 1: 3 → 64 channels (2 conv layers)
- Block 2: 64 → 128 channels (2 conv layers)
- Block 3: 128 → 256 channels (3 conv layers)
- Block 4: 256 → 512 channels (3 conv layers)
- Block 5: 512 → 512 channels (3 conv layers)

Progressive increase in channels allows the network to learn increasingly complex features, and early layers capture edges and textures, deeper layers capture landmark-specific patterns. Also, multiple conv layers per block allow for more complex feature combinations.

Next, I included BatchNorm after every convolutional layer. Including BatchNorm stabilizes training by normalizing inputs to each layer, and thus, reduces internal covariate shift as acting as a slight regularizer. 

Also, I used AdaptiveAvgPool2d((7, 7)) before the classifier. It makes the network flexible to different input sizes and reduces spatial dimensions while preserving feature information. This approach is often more robust than global average pooling for maintaining spatial relationships.

After feature extraction through convolutional blocks, I implemented a fully connected classifier to map high-level features to class predictions. The flattened output from the final convolutional block (after AdaptiveAvgPool2d((7, 7))) results in a 25,088-dimensional feature vector. This feeds into three linear (fully connected) layers:
- FC1: 25,088 → 4,096
- FC2: 4,096 → 4,096
- FC3: 4,096 → 50 (number of landmark classes)
By using two large hidden layers (4,096 units each), the model has sufficient capacity to learn complex and abstract feature interactions necessary for fine-grained classification. To mitigate overfitting due to the large number of parameters, I inserted Dropout(p=0.4) after each of the first two fully connected layers. This encourages the network to generalize better by randomly deactivating neurons during training.

Concurrently, several model designs were added to improve training stability, computational efficiency, and performance:
-  ReLU Activations after every convolutional and fully connected layer: Alleviates vanishing gradient issues, Encourages sparse activation, Fast and simple to compute
- MaxPool2d (2×2, stride=2) after each convolutional block: Reduces spatial dimensions, helping to control model size, Introduces translational invariance, Encourages focus on dominant features in each region
- No Skip Connections: Maintains simplicity in the initial design, Keeps the architecture VGG-like for easier debugging and interpretability, Makes focus remains on the core stacking of convolutions and activations

Last, the overall architecture was validated for both computational feasibility and baseline performance:
- Input resolution: 224×224
- Spatial dimension reduction across blocks: 224 → 112 → 56 → 28 → 14 → 7
- Receptive field increases progressively, allowing deeper layers to capture higher-level patterns
- BatchNorm and Dropout together provide regularization, helping prevent overfitting
- Achieved ~50% top-1 accuracy on a 50-class landmark classification task using a moderately sized dataset
